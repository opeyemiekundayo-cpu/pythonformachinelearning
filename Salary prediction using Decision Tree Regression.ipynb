```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

os.chdir('C:\\Users\\HP\\Desktop\\ML Course Content\\Part 2 - Regression\\Section 8 - Decision Tree Regression') # to change file working directory
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values # to convert all columns but dependent column to matrix to represent independent variables
y = dataset.iloc[:, 2].values
```

```python
#fitting the regression model
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

```python
# Predicting a new result
y_pred= regressor.predict(np.array(6.5).reshape(1,-1))
```

```python
y_pred
```

```python
# Visualizing the Decision Tree Regression model results for higher resolution
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Real or Bluff (Decision Tree Regression model)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualizing the Decision Tree Regression model results for higher resolution again
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Real or Bluff (Decision Tree Regression model)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
