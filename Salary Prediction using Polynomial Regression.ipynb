```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
os.chdir('C:\\Users\\HP\\Desktop\\ML Course Content\\Part 2 - Regression\\Section 6 - Polynomial Regression') # to change file working directory
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values # to convert all columns but dependent column to matrix to represent independent variables
y = dataset.iloc[:, 2].values
```

```python
X
```

```python
# fitting the multiple linear regression to dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# fitting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Visualizing the linear regression model
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Real or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualizing the polynomial regression model
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Real or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# increase the degree to 3
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Visualizing the polynomial regression model
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Real or Bluff ({Polynomial} Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# increase the degree to 4
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Visualizing the polynomial regression model
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Real or Bluff (Polynomialr Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#to make the line more refined
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
```

```python
# Visualizing the polynomial regression model with refined grid
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Real or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
